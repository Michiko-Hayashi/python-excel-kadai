```python
import openpyxl
from openpyxl.styles import Font
from datetime import datetime

# ワークブック作成
wb = openpyxl.Workbook()
# アクティブなシートを変数「ws」に格納
ws = wb.active
ws.title = '請求書'

# ヘッダー部分を作成
ws['B2'] = '請求書'
ws['B4'] = '株式会社ABC'
ws['B5'] = '〒101-0022 東京都千代田区神田錬塀町300'
ws['B6'] = 'TEL:03-1234-5678 FAX:03-1234-5678'
ws['B7'] = '担当者名：鈴木一郎様'
ws['F4'] = 'No.'
ws['G4'] = '0001'
ws['F5'] = '日付'
ws['G5'] = datetime.today().strftime('%Y/%m/%d')

# データを書き込む
headers = ['商品名', '数量', '単価', '金額']
for i, h in enumerate(headers):
    cell = ws.cell(row=10, column=2 + i)
    cell.value = h

data = [
    ['商品A', 2, 10000],
    ['商品B', 1, 15000],
]

start_row = 11
for r,(name, qty, unit) in enumerate(data, start=start_row):
    ws.cell(row=r, column=2, value=name)
    ws.cell(row=r, column=3, value=qty)
    ws.cell(row=r, column=4, value=unit)
    ws.cell(row=r, column=5, value=f"=C{r}*D{r}")
end_row = start_row + len(data) - 1

ws['B15'] = '合計'
ws['B16'] = '消費税'
ws['B17'] = '税込合計'

ws['E13'] = f"=SUM(E{start_row}:E{end_row})"
ws['E15'] = "=E13"
ws['E16'] = "=E15*0.1"
ws['E17'] = "=E15+E16"

#フォントを統一
uniform_font = Font(name='Calibri', size=10.5, bold=True)
for row in ws.iter_rows():
    for cell in row:
        if cell.value is not None:
        cell.font = uniform_font

#保存する
today = datetime.today()
filename = f"請求書_{today.strftime('%Y%m%d')}.xlsx"
wb.save(filename)
print(f'saved: {filename}')
```
