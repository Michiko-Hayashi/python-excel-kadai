```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill

file_list = ['2022_年間売上表.xlsx', '2023_年間売上表.xlsx']

# 結合後のデータを格納するための空のデータファイルを作成
df = pd.DataFrame()

# ファイルを一つずつ処理する
for file_name in file_list:
    # Excelファイルを読み込む
    temp_df = pd.read_excel(file_name)
    # データフレームの列名を指定
    temp_df.columns = ['売上年', '商品', '金額（千円）']
    # concat()でデータフレームを連結
    df = pd.concat([df, temp_df,], ignore_index=True)

# グループ化して合計金額を集約する
grouped = (
    df.groupby(['商品', '売上年'], as_index=False)['金額（千円）']
      .sum()
      .sort_values(['商品', '売上年'])
)

# Excelファイルの作成
writer = pd.ExcelWriter('売上集計表.xlsx')

# DateFrameオブジェクトをExcelファイルに書き込む
grouped.to_excel(writer, sheet_name='売上集計表', index=False)
writer.close()

print(grouped)

# ヘッダーの色を変更する
wb = load_workbook('売上集計表.xlsx')
ws = wb['売上集計表']
fill = PatternFill(patternType='solid', fgColor='F2F2F2')

for cell in ws[1]:
    cell.fill = fill
wb.save('売上集計表.xlsx')
print(grouped)
```
